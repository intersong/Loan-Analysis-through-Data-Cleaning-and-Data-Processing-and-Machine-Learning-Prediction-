```python
import pandas as pd
loans_2015 = pd.read_csv('LoanStats3d.csv', skiprows=1)
half_count = len(loans_2015) / 2
loans_2015 = loans_2015.dropna(thresh=half_count, axis=1)
loans_2015 = loans_2015.drop(['url'],axis=1)
loans_2015.to_csv('loans_2015.csv', index=False)
```

```python
loans_2015 = loans_2015.drop(["total_rec_int", "total_rec_late_fee", "recoveries", "collection_recovery_fee", "last_pymnt_d", "last_pymnt_amnt"], axis=1)
```

```python
import pandas as pd
loans_2015_filtered = pd.read_csv('loans_2015_cleaned_and_prepared.csv')


```

```python
print(loans_2015_filtered.info())
```

```python
cols = loans_2015_filtered.columns
train_cols = cols.drop("loan_status")
features = loans_2015_filtered[train_cols]
target = loans_2015_filtered["loan_status"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
# transform train data
X_train_scaled = scaler.transform(X_train)
# transform test data
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectPercentile, f_regression
from sklearn.model_selection import GridSearchCV
# change the precentile here to reduce the training
select = SelectPercentile(score_func=f_regression, percentile=5).fit(X_train_scaled, y_train)
X_selected = select.transform(X_train_scaled)
print(X_selected.shape)
```

```python
X_test_selected = select.transform(X_test_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_selected, y_train)
LR.score(X_test_selected, y_test)   
# much quicker than SVC and even higher scores
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_selected, y_train)
svc.score(X_test_selected, y_test)
# not higher than LogisticRegression ...
```

```python
# Then swith to test the RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, max_depth=9)
rf.fit(X_selected, y_train)
print(rf.score(X_selected, y_train))
print(rf.score(X_test_selected, y_test))
```

```python
rf1 = RandomForestClassifier(n_estimators=200, n_jobs=-1)
parameters_1 = {'max_features':['sqrt', 'log2'],'max_depth':[11, 13, 15]}   #New parameter set with higher 'max_depth'
grid_1 = GridSearchCV(rf1, parameters_1, cv=5)
grid_1.fit(X_selected, y_train)
grid_1.score(X_selected, y_train)
grid_1.score(X_test_selected, y_test)
```
